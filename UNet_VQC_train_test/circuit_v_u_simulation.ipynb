```python
################ Weiwen on 12-30-2020 ################
# Generate the circuit of u-layer
######################################################
#define your input and repeat times
u_layer = ULayerCircuit(4,2) 
#init circuit
circuit2 = QuantumCircuit()
#add input qubit to your circuit if needed
inps = u_layer.add_input_qubits(circuit2)

#add auxiliary qubit to your circuit if needed
aux =u_layer.add_aux(circuit2)

#add output qubit to your circuit if needed
u_layer_out_qubits = u_layer.add_out_qubits(circuit2)


#add ulayer to your circuit
unitary = torch.tensor(unitary,dtype=torch.double)
u_layer.forward(circuit2,binarize(weight),inps,u_layer_out_qubits,unitary,aux)

circuit2.barrier()

add_measure(circuit2,u_layer_out_qubits[0:2],'reg')


#show your circuit
if is_colab:
    layout = 'text'
    fold = 300
else:
    layout = 'mpl'
    fold = 50

circuit.draw(layout,fold=fold)

```
